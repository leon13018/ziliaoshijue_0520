站點流量統計

計算總流量
找出最大流量站點
印出超過50次的站名

```python
stations = [
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50}
]

for station in stations:
    print(station)
```

```python
# 站點流量統計

# 計算總流量
total = sum(s["count"] for s in stations)
print("總流量:", total)

# 找出最大流量站點
max_station = max(stations, key=lambda s: s["count"])
print("最大流量站點:", max_station["name"], "流量:", max_station["count"])

# 印出超過50次的站名
print("超過50次的站名:")
for s in stations:
    if s["count"] > 50:
        print(s["name"])
```
