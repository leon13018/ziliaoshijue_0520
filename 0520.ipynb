```python
# 載入所需套件：pandas 用於資料處理，json 用於解析 GeoJSON，os 用於檔案路徑操作
import pandas as pd
import json
import os
```

站點流量統計

計算總流量
找出最大流量站點
印出超過50次的站名

```python
# 定義站點清單，每筆資料包含站名（name）與流量次數（count）
stations = [
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50}
]

# 逐一印出每個站點的資料
for station in stations:
    print(station)
```

```python
# 計算所有站點的流量加總
total = sum(s["count"] for s in stations)
print("總流量:", total)

# 找出流量最大的站點（以 count 欄位為排序依據）
max_station = max(stations, key=lambda s: s["count"])
print("最大流量站點:", max_station["name"], "流量:", max_station["count"])

# 篩選並印出流量超過 50 次的站名
print("超過50次的站名:")
for s in stations:
    if s["count"] > 50:
        print(s["name"])
```

```python
# 指定存放日本各都道府縣 GeoJSON 地圖檔的資料夾路徑
folder = './Japanmap-master/prefs_064'

# 列出資料夾內所有檔案名稱
# print(os.listdir(folder))

# 初始化空串列，用來收集每個 feature 的資料列
rows = []

# 逐一讀取資料夾中所有 .geojson 檔案
for file in os.listdir(folder):
    if file.endswith('.geojson'):          # 只處理 GeoJSON 格式的檔案
        path = os.path.join(folder, file)  # 組合完整檔案路徑
        # print(path)
        with open(path, encoding='utf-8') as f:
            geo = json.load(f)             # 將 GeoJSON 檔案解析為 Python 字典
        # print(geo['type'])

        # 遍歷該檔案內的每個地理要素（feature）
        for feature in geo['features']:
            # print(feature['properties'])
            properties = feature['properties']  # 取出屬性資料（如縣名）
            geometry = feature['geometry']      # 取出幾何資料（如邊界座標）

            # 將本筆資料整理成字典，包含檔名、縣碼前綴、縣名、屬性
            row = {
                'file_name': file,
                'pref_code': file[:2],                    # 取檔名前兩字母作為縣碼
                'pref_name': file.replace('.geojson', ''), # 去除副檔名得到縣名
                'properties': properties
            }
            rows.append(row)  # 將該筆資料加入串列

# 將所有資料列轉換為 DataFrame 並顯示
pd.DataFrame(rows)
```
