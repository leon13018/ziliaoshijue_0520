站點流量統計

計算總流量
找出最大流量站點
印出超過50次的站名

```python
# 定義站點清單，每筆資料包含站名（name）與流量次數（count）
stations = [
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50},
    {'name':'凹子底', 'count':50}
]

# 逐一印出每個站點的資料
for station in stations:
    print(station)
```

```python
# 計算所有站點的流量加總
total = sum(s["count"] for s in stations)
print("總流量:", total)

# 找出流量最大的站點（以 count 欄位為排序依據）
max_station = max(stations, key=lambda s: s["count"])
print("最大流量站點:", max_station["name"], "流量:", max_station["count"])

# 篩選並印出流量超過 50 次的站名
print("超過50次的站名:")
for s in stations:
    if s["count"] > 50:
        print(s["name"])
```
